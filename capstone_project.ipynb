```python
sns.boxplot(x='hotel', y='lead_time', data=df)
```

The boxplot shows that Resort Hotel bookings have significantly higher lead times than City Hotel bookings. This suggests that resort stays are generally planned further in advance, likely due to vacation and leisure travel, whereas city hotel bookings are made closer to the arrival date, possibly for business or short-notice travel.
