```python
df.groupby('assigned_room_type')['adr'].mean().sort_values(ascending=False) #H generaees highest revenue
```

```python
df.groupby('assigned_room_type')['adr'].mean().plot(kind='bar')#through barplot we se which room type earns most
```

```python
sns.boxplot(x='assigned_room_type', y='adr', data=df)#through boxpot we are checking is pricing stable or volatile?
```

The boxplot and barplot use the same variables but serve different purposes. The barplot compares average ADR across room types, while the boxplot analyzes ADR distribution, spread, median, and outliers within each room category.

Higher room types such as G, H, and F command premium pricing. Lower room types like L and P generate lower ADR. Some room types show wide pricing variation and outliers, indicating dynamic pricing or anomalies.

This helps optimize room pricing, promotions, and inventory mix.

```python
df.groupby('market_segment')['adr'].mean().sort_values(ascending=False) #to find high-value customers
```

```python
df.groupby('market_segment')['adr'].mean().plot(kind='bar')
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x='market_segment', y='adr', data=df)
```

Barplot compares average ADR across booking segments, while boxplot reveals pricing spread, median, and outliers within each segment.

Direct and Online TA segments generate the highest ADR, making them premium revenue channels. Corporate bookings have lower ADR due to negotiated rates. Complimentary bookings generate minimal revenue.

Hotels can prioritize profitable acquisition channels and optimize channel mix
